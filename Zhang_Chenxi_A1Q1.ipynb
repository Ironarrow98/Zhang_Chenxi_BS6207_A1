```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import torch
import torch.nn.functional as F
from torch.autograd import Variable
```

# Network

```python
K = 10
d = 10


net = torch.nn.Sequential(
    # Input Layer
    torch.nn.Linear(10, 21),
    # 1st Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 2nd Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 3rd Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 4th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 5th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 6th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 7th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 8th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 9th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # 10th Layer
    torch.nn.Linear(21, 21),
    torch.nn.ReLU(),
    # Output Layer
    torch.nn.Linear(21, 1)
)

print(net)
```

# Sample X

```python
runif = torch.distributions.Uniform(0, 1)
X = runif.sample((1, 10))[0]
X
```

# Sample y

```python
y = 0
for x in X:
    y += (x * x)
y /= d
y
```

# Loss Function

```python
def Loss(predict, y):
    return (predict - y) ** 2
```

# Autograd Forward and Backward

```python
prediction = net(X)
loss = Loss(prediction, y)
loss.backward()
print("Loss = " + str(loss.data.numpy()[0]))
```

```python
input_layer = net[0]
print("Loss = " + str(np.round(loss.item(), 5)))
print("y_prediction: " + str(np.round(prediction.tolist()[0], 5)))
print("\n")
print("Input_layer weight gradient: \n" + str(np.round(input_layer.weight.grad.tolist(), 5)))
print("\n")
print("Input_layer bias gradient: \n" + str(np.round(input_layer.bias.grad.tolist(), 5)))
print("\n")

with open('torch_autograd.dat', 'a') as the_file:
    for i in range(1, 22, 2):
        the_file.write("Hidden_layer " + str((i + 1) // 2) + " weight gradient: \n" + str(np.round(net[i].weight.grad.tolist(), 5)))
        the_file.write("\n")
        print("Hidden_layer " + str((i + 1) // 2) + " weight gradient: \n" + str(np.round(net[i].weight.grad.tolist(), 5)))
        print("\n")
        the_file.write("Hidden_layer " + str((i + 1) // 2) + " bias gradient: \n" + str(np.round(net[i].bias.grad.tolist(), 5)))
        the_file.write("\n")
        print("Hidden_layer " + str((i + 1) // 2) + " bias gradient: \n" + str(np.round(net[i].bias.grad.tolist(), 5)))
        print("\n")
```

# Manual Forward

```python
def compute_output(x, w, b):
    return x.dot(w) + b

weights = []
bias = []
for i in range(22):
    if type(net[i]) != torch.nn.modules.activation.ReLU: 
        weights.append(net[i].weight.t().detach().numpy())
        bias.append(net[i].bias.detach().numpy())
    
y_pred = X.numpy()
hiddens = []
for i in range(len(weights)):
    cur_out = compute_output(y_pred, weights[i], bias[i])
    y_out = torch.from_numpy(cur_out)
    if i != 0 and i != 11:
        m = torch.nn.ReLU()
        y_pred = m(y_out).numpy()
    else: 
        y_pred = y_out.numpy()
    hiddens.append(y_pred)

loss_forward = ((y_pred - y.numpy()) ** 2)[0]
print("Loss = " + str(loss_forward))
```

# Manual Backward

```python
def d_relu(x):
    x[x <= 0] = 0
    x[x > 0] = 1
    return x

dt = float(2 * (y_pred - y.numpy()))

# Output Layer
dw_out = np.dot(hiddens[-2].T, dt)
db_out = np.ones(1).dot(dt)

# Hidden Layer
i = len(weights) - 2
dws = []
dbs = []
while i >= 1: 
    h = hiddens[i].copy()
    d = d_relu(h)
    dt = np.dot(dt, weights[i + 1].T) * d
    db = np.ones(1).dot(dt)
    dw = np.dot(hiddens[i - 1].reshape(len(hiddens[i - 1]), 1), dt)
    dws.insert(0, dw)
    dbs.insert(0, db)
    i -= 1

# Input Layer
dt = np.dot(dt, weights[1].T) * d_relu(hiddens[0].copy())
dw_in = np.dot(X.numpy().reshape(len(X.numpy()), 1), dt)
db_in = np.ones(1).dot(dt)

print("Loss = " + str(round(loss_forward, 5)))
print("y prediction: " + str(np.round(y_pred, 5)))
print("\n")
print("Input_layer weight gradient: " + str(np.round(dw_in.T, 5)))
print("\n")
print("Input_layer bias gradient: " +  str(np.round(db_in, 5)))
print("\n")

with open('my_autograd.dat', 'a') as the_file:
    for i in range(len(dbs)):
        the_file.write("Hidden_layer " + str(i + 1) + " weight gradient: \n" + str(np.round(dws[i], 5)))
        the_file.write("\n")
        print("Hidden_layer " + str(i + 1) + " weight gradient: \n" + str(np.round(dws[i], 5)))
        print("\n")
        the_file.write("Hidden_layer " + str(i + 1) + " bias gradient: \n" + str(np.round(dbs[i], 5)))
        the_file.write("\n")
        print("Hidden_layer " + str(i + 1) + " bias gradient: \n" + str(np.round(dbs[i], 5)))
        print("\n")
    
    the_file.write("Output_layer weight gradient: " + str(np.round(dw_out.T, 5)))
    the_file.write("\n")
    print("Output_layer weight gradient: " + str(np.round(dw_out.T, 5)))
    print("\n")
    the_file.write("Output_layer bias gradient: " +  str(np.round(db_out, 5)))
    the_file.write("\n")
    print("Output_layer bias gradient: " +  str(np.round(db_out, 5)))
    print("\n")
```
